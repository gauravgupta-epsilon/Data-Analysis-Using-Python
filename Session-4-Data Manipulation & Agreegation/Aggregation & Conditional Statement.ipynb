#### Create basic fumction in Python

```python
df.head()
```

```python
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
```

```python
df['Gender_N'] = df.gender.apply(gender_to_numeric)
```

```python
df1 = df.reset_index()
```

```python
pd.DataFrame(df.groupby(['occupation']).Gender_N.sum()/df.groupby(['occupation']).Gender_N.count())
```

#### Lambda function in Python.

```python
df['Gen_count'] = df.gender.apply(lambda x: 1 if (x == 'M') else 0)
```

```python
# create a data frame and apply count to gender
gender_ocup = df.groupby(['occupation', 'gender']).agg({'Gender_N': 'count','Gender_N': 'sum'})
gender_ocup.head()
```

```python
gender_ocup.reset_index()
```

#### Pivot Table in Python.

```python
# Calculate the Ration of Male & female in each group with each Method..
# Calculate the % of Male & female in each group with All Methods..
```

```python
df_pivot = (df.pivot_table(index=['occupation'], 
                      columns='gender', 
                      values='age', 
                      aggfunc='count')
         .reset_index()
         .rename_axis(None, axis=1))
```

```python
df_pivot.head()
```
