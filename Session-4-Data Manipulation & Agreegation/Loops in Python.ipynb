#### | Index attribute of the Dataframe.

```python
date1 = pd.Timestamp.now()
curr_year = date1.year
curr_year
```

```python
for i in df.index:
    i = i+1;
    yob = curr_year - df['age'][i]
    df['YOB'][i] = yob
    print(yob)
```

#### | loc[] function of the Dataframe.

```python
for i in df.index:
    i = i+1;
    yob = curr_year - df.loc[i,'age']
    df.loc[i,'YOB'] = yob
    print(yob)
```
