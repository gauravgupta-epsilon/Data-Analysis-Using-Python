# Filtering/Subsetting Data

### Filtering Columns

```python
#using column names
df_store1 = df_store[['STORE_ID', 'STORE_NAME', 'ADDRESS_CITY_NAME']]
df_store1
```

```python
#using loc function
df_store2 = df_store.loc[:,('STORE_ID', 'STORE_NAME', 'ADDRESS_CITY_NAME')]
df_store2
```

```python
#using iloc function
df_store3 = df_store.iloc[:,:3]
df_store3
```

### Filtering Rows

```python
#using loc function
df_store1 = df_store.loc[:10,("STORE_ID","STORE_NAME")]
df_store1
```

```python
#using iloc function
df_store2 = df_store.iloc[:10,:2]
df_store2
```

### Filtering dataframe based on condition

```python
##Filter all the stores which are located in the city HOUSTON
df_store_city = df_store[df_store['ADDRESS_CITY_NAME']=='HOUSTON']
df_store_city
```

```python
##isin keyword
##Filter all the stores which are located in the city HOUSTON or DALLAS
df_store_city = df_store[df_store['ADDRESS_CITY_NAME'].isin(['HOUSTON','DALLAS'])]
#df_store_city
```

```python
##~ operator
##Filter all the stores which are not located in the city HOUSTON or DALLAS
df_store_city = df_store[~df_store['ADDRESS_CITY_NAME'].isin(['HOUSTON','DALLAS'])]
df_store_city
```

```python
##Filter all the stores which are located in the city HOUSTON using loc function
df_store_city = df_store.loc[df_store['ADDRESS_CITY_NAME']=='HOUSTON',:]
df_store_city
```

```python
##Applying multiple condions for filtering
a = df_store[(df_store['ADDRESS_STATE_PROV_CODE']=='TX') |  (df_store['SALES_AREA_SIZE_NUM']>=20000)]
a
```

### Adding a new column to dataframe

```python
df_store_city['AVG_MONTHLY_BASKETS'] = df_store_city['AVG_WEEKLY_BASKETS']*4
```

```python
df_store_city
```

### Applying If else condions in multiple ways in pandas

```python
#Create new column based on certain condition
df_store_city['Houston_city_flag'] = np.where(df_store_city['ADDRESS_CITY_NAME']=='HOUSTON',1,0)
```

```python
df_store_city
```

```python
#one more way of creating a new column based on condition
 
df_store.loc[df_store['ADDRESS_STATE_PROV_CODE']=='OH', 'OH_STATE_FLAG'] = 1
df_store.loc[df_store['ADDRESS_STATE_PROV_CODE']!='OH', 'OH_STATE_FLAG'] = 0
df_store
```

```python
#Apply if condition using lambda

df_store['TX_STATE_FLAG'] = df_store['ADDRESS_STATE_PROV_CODE'].apply(lambda x: 1 if x=='TX' else 0)
df_store
```
